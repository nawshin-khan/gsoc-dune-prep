## Task 2 - Plot Temperature vs. Time for a Cooling Process
**Task:** Simulate a cooling process using $T(t) = T_0 e^{-kt}$, where $T_0 = 100$, $k = 0.05$, and $t$ ranges from 0 to 100.

**Requirements:**
- Generate the data programmatically.
- Plot $T(t)$ with appropriate labels and a grid.

```python
def model(T0,t):
    T=T0*np.exp(-k*t)
    return T
T0=100
k=0.05
t=np.linspace(0,100)
T=model(T0,t)
plt.plot(t,T)
plt.grid(True)
plt.xlim(0,100)
plt.xlabel("t")
plt.ylabel("$T(t) = T_0 e^{-kt}$")

plt.legend()
plt.show()
```

## Task 3 - Compare Damped Oscillations
**Task:** Plot damped oscillation curves $y(t) = A e^{-bt} \cos(\omega t)$ for different damping coefficients $b$.

**Requirements:**
- Plot 3 curves with $b = 0.1, 0.5, 1$, keeping $A = 1$ and $\omega = 2\pi$.
- Label the curves with the corresponding $b$ values.

```python
def model(t,b):
    
    y=A*np.exp(-b*t)*np.cos(omega*t)
    return y

A=1
omega=2*np.pi
t=np.linspace(0,20)
b=.1
y1=model(t,b)
b=.5
y2=model(t,b)
b=1
y3=model(t,b)

plt.plot(t,y1,'r-',label="b=.1")
plt.plot(t,y2,'b--',label="b=.5")
plt.plot(t,y3,'g:',label="b=1")
plt.xlabel("t")
plt.ylabel("$y(t) = A e^{-bt} \cos(\omega t)$")
plt.legend()
plt.grid(True)
plt.show()
```

## Task 8 - Simulate and Plot Projectile Motion
**Task:** Simulate a projectile motion path for a given initial velocity and angle.

**Requirements:**
- Plot the trajectory $y(x)$.
- Add markers at key points like maximum height and range.

```python
v0=30
theta=np.radians(45)
g=9.8

t=np.linspace(0,2*v0*np.sin(theta)/g)
x=v0*np.cos(theta)*t
y=v0*np.sin(theta)*t-0.5*g*t**2

tp=v0*np.sin(theta)/g
xp=v0*np.cos(theta)*tp
yp=v0*np.sin(theta)*tp-0.5*g*tp**2

plt.plot(x,y,label="Trajectory")

plt.plot(xp,yp,'ro',label="Max Height")
plt.plot(x[-1],0,'bo',label="Range")

plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.grid(True)
plt.legend()
plt.show()
```

## Task 10 - Plot Resistance as a Function of Temperature
**Task:** Generate resistance values using $R(T) = R_0(1 + \alpha T)$, where $R_0 = 10$, $\alpha = 0.005$, and $T$ ranges from 0 to 100.

**Requirements:**
- Plot $R(T)$ with proper labels and a legend.

```python
R0=10
alpha=.005
def model(T):
    R=R0*(1+ alpha*T)
    return R

T=np.linspace(0,100)
R=model(T)
plt.plot(T,R,label="$R(T) = R_0(1 + \\alpha T)$")
plt.grid(True)
plt.xlim(0,100)
plt.xlabel("T")
plt.ylabel("$ R(T)= R_0(1+ \\alpha T)$")


plt.legend()
plt.show()
```

## Task 11 - Plot Energy Bands in a Material
**Task:** Simulate the energy bands of a material using the dispersion relation $E(k) = \frac{\hbar^2 k^2}{2m}$.

**Requirements:**
- Plot energy vs. wave vector $k$ for multiple bands with different effective masses $m$.

```python
def model(k,m):
    E=(hbar**2*k**2)/2*m
    return E
hbar= 1.055e-34
em= 9.11e-31
k=np.linspace(0,20)
m=.5*em
E1=model(k,m)
m=em
E2=model(k,m)
m=2*em
E3=model(k,m)

plt.plot(k,E1,'r-',label="$m=.5e_m$")
plt.plot(k,E2,'b--',label="$m=e_m$")
plt.plot(k,E3,'g:',label="$m=2e_m$")
plt.xlabel("k")
plt.ylabel("$E_k = \\frac{\\hbar^2 k^2}{2m}$")
plt.legend()
plt.grid(True)
plt.show()
```
